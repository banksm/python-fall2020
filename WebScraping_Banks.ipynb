```python
import requests

response = requests.get("https://www.theverge.com/tech")
```

```python
import bs4
soup = bs4.BeautifulSoup(response.text, 'html.parser')
```

```python
soup
```

```python
soup.select(".c-entry-box-base__headline")
```

```python
headlines = []
for item in soup.select(".c-entry-box-base__headline"):
    headlines.append(item.text)
```
